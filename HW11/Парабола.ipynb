```python
from sympy import *
import matplotlib.pyplot as plt
import matplotlib as mpl
from sympy.plotting import plot
```

```python
x = symbols('x')
f = 5 * x ** 2 + 10 * x - 30
f
```

```python
roots = solve(f, x)
roots
```

```python
interv = Interval(float(roots[1]), float(roots[0]))
res_min = minimum(f, x, interv)
res_min
```
