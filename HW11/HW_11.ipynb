```python
# f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
# Определить корни
# Найти интервалы, на которых функция возрастает
# Найти интервалы, на которых функция убывает
# Построить график
# Вычислить вершину
# Определить промежутки, на котором f > 0
# Определить промежутки, на котором f < 0
```

```python
import math
from sympy import *
from sympy.plotting import plot as plt
from sympy import Symbol
from sympy import minimum, maximum, Interval
```

```python
x = symbols('x')
f = (-12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30)
f
```

```python
# Построить график
plt(f, (x,-100,100))
```

```python
# Определить корни
roots = solve(f)

# Расчет корней приводит к зависанию, потому что имеется бесконечное количество корней.

# Функция тригонометрическая, можно посчитать параметры только для определенных интервалов этой функции.
# Для параболы все параметры легко считаются.
```

```python
# Найти интервалы, на которых функция возрастает
solve(diff(f)>0)
# Найти интервалы, на которых функция убывает
solve(diff(fx)<0)

# Расчет интервалов приводит к зависанию, потому что имеется бесконечное количество интервалов.
```

```python
# Вычислить вершину
solve(diff(f))

# Расчет приводит к зависанию, т.к. вершина бесконечна.
```

```python
# Определить промежутки, на котором f > 0
print(solve(f>0))

# Определить промежутки, на котором f < 0
print(solve(f<0))

# Расчет интервалов приводит к зависанию, потому что имеется бесконечное количество промежутков.
```
